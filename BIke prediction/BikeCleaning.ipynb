### Supervised MAchine LEarning

Supervised Learning is a type of machine learning in which a model learns from labeled data. Each training example consists of:

Input (features) → the data provided to the model
Output (label/target) → the correct answer associated with that input

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42
)
```

```python
x_train.shape, y_train.shape , x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
bike_1 = lr.predict([[16329.0	,1,	4.0,	180.0,	11]])
bike_1
```

```python
print(lr.score(x_train,y_train))
print(lr.score(x_test,y_test))
```

```python
import joblib
joblib.dump(lr , 'model.lb')
```
