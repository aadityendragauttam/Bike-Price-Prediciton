### Machine Learning

* Machine Learning is a method of teaching computers to learn from data and improve their performance automatically.


| Type                     | Description                          |
| ------------------------ | ------------------------------------ |
| Supervised Learning      | Learns from labeled data             |
| Unsupervised Learning    | Learns from unlabeled data           |
| Semi-Supervised Learning | Uses both labeled and unlabeled data |
| Reinforcement Learning   | Learns through rewards and penalties | 




| Labeled Data                 | Unlabeled Data                   |
| ---------------------------- | -------------------------------- |
| Has correct answers (labels) | No correct answers               |
| Used in Supervised Learning  | Used in Unsupervised Learning    |
| Easier to train models       | Used to discover hidden patterns |
| Example: House Price Dataset | Example: Customer Purchase Data  |

| Aspect            | Classification                                                           | Regression                                                                  |
| ----------------- | ------------------------------------------------------------------------ | --------------------------------------------------------------------------- |
| Output            | A category or class                                                      | A continuous numerical value                                                |
| Goal              | Assign data to predefined labels                                         | Predict a quantity                                                          |

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
x = df.drop('price',axis=1)
```

```python
y = df[['price']]
```

```python
x
```

```python
y
```

```python
#pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape , y_train.shape , x_test.shape , y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()     #object creation
lr.fit(x_train,y_train)
```

```python
print(lr.score(x_train,y_train))
print(lr.score(x_test,y_test))
```
